This code is built to propagate labels on biological networks. The label propagation algorithm was developed and modeled by the [HotNet Algorithm](https://www.researchgate.net/figure/Overview-of-the-HotNet-algorithm-HotNet-102-uses-a-heat-diffusion-process-to-identify_fig4_259987271). This algorithm is used to identify genes that connect to the development of cancer and how these abnormalities spread through the genome. 

**This code is split into three different sections**
* The creation and testing of the label propagation algorithm
* The label propagation algorithm being used to "recover" seed nodes
* The comparison of the model to the [K-means clustering](https://en.wikipedia.org/wiki/K-means_clustering) algorithm

### Note about Color Maps:
> "No relation" means that the node has no relation to the abnormality\
> "Related" indicates the node shows some affect or connection to the abnormality\
> "Highly Related" shows that the node is a part of the abnormality

 ## Label Propagation Algorithm

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import graphviz_layout
from sklearn.cluster import KMeans
import pandas as pd

import warnings
warnings.filterwarnings(action='ignore')

def label_propagation(graph, seed_nodes, constant):
    
    # Making labels
    labels = np.zeros(len(graph))
    for x in range(len(seed_nodes)):
        labels[seed_nodes[x]] = constant[x]
    
    fst = np.transpose(np.array(labels))

    #adjacency matrix should technically only be 1 or 0
    degree = np.diag(np.sum(graph, axis=1))
    laplacian = degree - graph

    constant_diag = np.eye(len(graph))
    i=0
    for seed in seed_nodes:
        constant_diag[seed, seed] = constants[i]
        i+=1
    Ly = laplacian + constant_diag

    bs = np.zeros(len(graph))
    for seed in seed_nodes:
        bs[seed] = 1

    if np.linalg.det(Ly) == 0:
        # Sometimes the matrix can be singular, so we estimate inverse with pinv
        fs = np.matmul(-1*np.linalg.pinv(Ly), bs)
    else:
        fs = np.matmul(-1*np.linalg.inv(Ly), bs)
    #bs is the nodes with a known start (1 for known, 0 otherwise)

    return fs

# Create a scale free graph
G = nx.scale_free_graph(n=20, alpha=0.31, beta=0.64, gamma=0.05, seed=150)

# Seed nodes with known labels
seed_nodes = [0, 5, 10, 15]
constants = [-1, 1, 1, 1] # Negative 1 indicates no relationship, Positive 1 indicates connection

# Example usage
labels = np.zeros(len(G))
    
for x in range(len(seed_nodes)):
    labels[seed_nodes[x]] = constants[x]
    

# Run label propagation
final_labels = label_propagation(nx.to_numpy_array(G), seed_nodes, constants)

# Print the final labels
final_labels
```

#### Testing/Plotting

```python
pos = nx.spring_layout(G, seed=150)
plt.figure(figsize=(12, 10))

# Plot original
plt.subplot(2, 2, 1)
nx.draw(G, pos, node_size=100, edge_color='black', node_color=labels, cmap='autumn_r', with_labels=True)
plt.title("Original Network")

# Plot propogated
plt.subplot(2, 2, 2)
nx.draw(G, pos, node_size=100, edge_color = 'black', node_color=final_labels, cmap='autumn_r', with_labels=True)
plt.title("Network with Propagated Labels")

sm = plt.cm.ScalarMappable(cmap='autumn_r')
sm._A = []
cbar = plt.colorbar(sm, cax = plt.axes([0.5, 0.5, 0.01, 0.35]))
cbar.set_ticks([0.0, 0.5, 1.0])  # Example: Set custom tick positions (values between 0 and 1)
cbar.set_ticklabels(["No Relation", "Related", "Highly Related"]) # relation to cancer

# Plot difference
plt.subplot(2, 2, 3)
nx.draw(G, pos, node_size=100, edge_color = 'black', node_color=abs(labels-final_labels), cmap='Blues', with_labels=True)
plt.title("Label Propagation Difference"); # the liquid leaves the seed nodes

sm = plt.cm.ScalarMappable(cmap='Blues')
sm._A = []
cbar = plt.colorbar(sm, cax = plt.axes([0.5, 0.1, 0.01, 0.35]))
cbar.set_ticks([0.0, 0.5, 1.0])  # Example: Set custom tick positions (values between 0 and 1)
cbar.set_ticklabels(["Same", "Similar", "Different"])

plt.show()
```

### Recovering Seed Nodes

```python
# Create a scale free graph
G = nx.scale_free_graph(n=50, alpha=0.31, beta=0.64, gamma=0.05, seed=150)

# Seed nodes with known labels
seed_nodes = [0, 5, 10, 15, 20, 30, 35, 40, 45]
constants = [1, 1, 1, -1, 1, 1, 1, -1, 1]    

# Run label propagation
original_labels = label_propagation(nx.to_numpy_array(G), seed_nodes, constants)

# NOW REMOVE SOME SEED NODES - reduction of 1/3
# Seed nodes with known labels
seed_nodes = [0, 5, 15, 20, 35, 45]
constants = [1, 1, -1, 1, 1, 1]    

# Run label propagation
recovered_labels = label_propagation(nx.to_numpy_array(G), seed_nodes, constants)
```

Overall, as long as you remove seed nodes in a manner that is evenly spaced, the algorithm can handle a 1/3 reduction of seed nodes with little to no issues.

#### Plotting

```python
pos = nx.spring_layout(G, seed=150)
plt.figure(figsize=(12, 10))

# Plot original
plt.subplot(2, 2, 1)
nx.draw(G, pos, node_size=40, edge_color='black', node_color=original_labels, cmap='autumn_r', with_labels=False)
plt.title("Original Propagation")

# Plot propogated
plt.subplot(2, 2, 2)
nx.draw(G, pos, node_size=40, edge_color = 'black', node_color=recovered_labels, cmap='autumn_r', with_labels=False)
plt.title("Network with Recovered Labels")

sm = plt.cm.ScalarMappable(cmap='autumn_r')
sm._A = []
cbar = plt.colorbar(sm, cax = plt.axes([0.5, 0.5, 0.01, 0.35]))
cbar.set_ticks([0.0, 0.5, 1.0])  # Example: Set custom tick positions (values between 0 and 1)
cbar.set_ticklabels(["No Relation", "Related", "Highly Related"]) # relation to cancer

# Plot difference
plt.subplot(2, 2, 3)
nx.draw(G, pos, node_size=40, edge_color = 'black', node_color=abs(original_labels-recovered_labels), cmap='Blues', with_labels=False)
plt.title("Label Propagation Difference"); # the liquid leaves the seed nodes

sm = plt.cm.ScalarMappable(cmap='Blues')
sm._A = []
cbar = plt.colorbar(sm, cax = plt.axes([0.5, 0.1, 0.01, 0.35]))
cbar.set_ticks([0.0, 0.5, 1.0])  # Example: Set custom tick positions (values between 0 and 1)
cbar.set_ticklabels(["Same", "Similar", "Different"])

plt.show()
```

### K-means Clustering Comparison

```python
# Create a scale free graph
G = nx.scale_free_graph(n=20, alpha=0.31, beta=0.64, gamma=0.05, seed=150)

pos = graphviz_layout(G, prog='neato')

# Coordinate representation of graph
graph = pd.DataFrame(np.matrix([pos[i] for i in pos.keys()]))

graph['final_labels'] = final_labels
```

```python
kmeans = KMeans(n_clusters=int(len(np.unique(labels))))
kmeans.fit(np.array(graph))
final_labels = kmeans.labels_

print("Final Labels:")
print(final_labels)
```

```python
pos = nx.spring_layout(G, seed=150)
plt.figure(figsize=(12, 6))

# Plot original
plt.subplot(1, 2, 1)
nx.draw(G, pos, node_size=100, edge_color='black', node_color=labels, cmap='autumn', with_labels=True)
plt.title("Original Network")

# Plot propogated
plt.subplot(1, 2, 2)
nx.draw(G, pos, node_size=100, edge_color = 'black', node_color=final_labels, cmap='cool', with_labels=True)
plt.title("Network with KMeans Groups");
```
