```python
import networkx as nx
import matplotlib.pyplot as plt

# Generate a scale-free directed graph
G = nx.scale_free_graph(n=20, alpha=0.41, beta=0.54, gamma=0.05, delta_in=0.05, delta_out=0, seed=10374196)

# Visualize the graph
pos = nx.spring_layout(G, seed=10396953)
nx.draw(G, pos, with_labels=True, node_size=200, node_color="skyblue", font_size=8, font_color="black", font_weight="bold")
plt.title("Scale-Free Network")
plt.show()
plt.savefig("scalefree.png")
```

```python
import numpy as np

# Convert the graph to an adjacency matrix
adj_matrix = nx.to_numpy_array(G, dtype=int)
adj_matrix.shape
```
